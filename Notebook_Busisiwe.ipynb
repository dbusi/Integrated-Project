```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance_claims.csv")
```

```python
df.head()
```

```python
null_counts = df.isnull().sum()
null_counts
```

```python
df_new = df.drop('_c39', axis=1, errors='ignore')
```

```python
df_new.head()
```

```python
df_new.describe()
```

```python
# Choose the column for the histogram
column_name = 'age'

# Plot the histogram
plt.hist(df[column_name], bins=3, edgecolor='black')

# Add labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

# Display the histogram
plt.show()
```

```python
bin_edges = [0, 30, 55, 100]  # Define the bin edges
bin_labels = ['Young Adult', 'Middle Aged', 'Elderly']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['ages_category'] = pd.cut(df_new['age'], bins=bin_edges, labels=bin_labels)
```

```python
df_new.head()
```

```python
bin_edges_customer = [0, 25, 150, 500]  # Define the bin edges
bin_labels_customer = ['New Client', 'Established Client', 'Long-Term Client']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['customer_category'] = pd.cut(df_new['months_as_customer'], bins=bin_edges_customer, labels=bin_labels_customer)
```

```python
df_new.head()
```

```python
df_new["Contract Years"] = df_new["months_as_customer"]/12
```

```python
df_new.head()
```

```python
df_new['total_premiums_paid'] = (df_new['policy_annual_premium']/12) * df_new['months_as_customer']
```

```python
df_new.head()
```

```python
df_new['net_value_of_customer'] = df_new['total_premiums_paid'] - df_new['total_claim_amount']
```

```python
df_new.head()
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x='auto_make', y='total_claim_amount', data=df_new, palette='Set3')
plt.title('Vehicle Make vs. Total Claim Amount')
plt.xlabel('Auto Make')
plt.ylabel('Total Claim Amount')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df_new['claim_ratio'] = df_new['total_claim_amount'] / df_new['policy_annual_premium']

```

```python
df_new.head()
```

```python
df_new['incident_month'] = pd.to_datetime(df_new['incident_date']).dt.month
plt.figure(figsize=(12, 8))
sns.countplot(x='incident_month', data=df_new, palette='Set1', hue='fraud_reported')
plt.title('Month-wise Incident Distribution')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(df_new.corr(), cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='umbrella_limit', data=df_new, palette='coolwarm', hue='fraud_reported')
plt.title('Umbrella Limit vs. Fraud Reported')
plt.xlabel('Umbrella Limit')
plt.ylabel('Count')
plt.show()
```

```python
df_new['annual_premium_per_vehicle'] = df_new['policy_annual_premium'] / df_new['number_of_vehicles_involved']
```

```python
df_new.head()
```

```python
avg_age = df_new['age'].mean()
df_new['age_diff_from_avg'] = np.abs(df_new['age'] - avg_age)
```

```python
df_new.head()
```

```python
df_new['previous_claim'] = (df_new['total_claim_amount'] > 0).astype(int)
```

```python
df_new.head()
```

```python
education_claim_avg = df_new.groupby('insured_education_level')['total_claim_amount'].mean().to_dict()
df_new['education_claim_avg'] = df_new['insured_education_level'].map(education_claim_avg)
```

```python
df_new.head()
```

```python
plt.figure(figsize=(8, 5))
df_new['previous_claim'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Proportion of Customers with Previous Claims')
plt.ylabel('')
plt.show()


```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='incident_severity', data=df_new, palette='viridis')
plt.xlabel('Incident Severity')
plt.ylabel('Count')
plt.title('Distribution of Incident Severity')
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='bodily_injuries', y='total_claim_amount', data=df_new, color='purple')
plt.xlabel('Number of Bodily Injuries')
plt.ylabel('Total Claim Amount')
plt.title('Bodily Injuries vs. Total Claim Amount')
plt.show()
```
